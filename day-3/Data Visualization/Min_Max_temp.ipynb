```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('min_max_temp.csv')
```

```python
data.set_index('Date', inplace=True)
```

```python
data
```

```python
max_temps = data['Tmax']
min_temps = data['Tmin']
```

```python
plt.figure(color='orange',figsize=[7,4])
plt.plot(max_temps)
```

```python
plt.figure(color='red',figsize=[7,4])
plt.plot(min_temps)
```
